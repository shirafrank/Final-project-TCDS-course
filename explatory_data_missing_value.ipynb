# # Adding new column - union of days since the prior order to weeks. "days_previous_Group"
Division of days since previous purchase by groups and adding to new column

```python
HGroups_days = {'1':list(range(0,8)),'2':list(range(8,14)),'3':list(range(14,21)),'4':list(range(21,31)) }
#Weekdays = {0:'Monday',1:'Tuesday',2:'Wednesday',3:'Thursday',4:'Friday',5:'Saturday',6:'Sunday'}v
```

```python

def hourgroup(a):
    for key, values in HGroups_days.items():
        if a in values:
            return key
```

```python
df['days_previous_Group'] = df.days_since_prior_order.apply(hourgroup)
```
