# Preparing the partition to test, train and dev

```python
import matplotlib.pyplot as plt
#%matplotlib inline
```

```python
from random import seed, shuffle

seed(200)
xind = [i for i in range(len(df))]
print(xind[0:15])
shuffle(xind,)
print(xind[0:15])
```

```python
### We will split as follows: 10% test, 10% dev and 80% train
### test split
split_1 = int(0.1 * len(xind))
### train-dev split
split_2 = int(0.2 * len(xind))
```


## Proceeding with the partition

```python
### We will split as follows: 10% test, 10% dev and 80% train
X_test = vx.iloc[xind[0:split_1],]
y_test = vy.iloc[xind[0:split_1],]

X_dev = vx.iloc[xind[split_1:split_2],]
y_dev = vy.iloc[xind[split_1:split_2],]

X_train = vx.iloc[xind[split_2:],]
y_train = vy.iloc[xind[split_2:],]
```
